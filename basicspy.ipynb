```python
''' four built in datatypes of python
1}list
2}tupple
3}dict
4}set'''


# LIST
#1]  mutable sequence of values
#2] we can store diff type of values
#3] can access values from index

marks=[99,100,20,30,45,26,10,56,89,"as"]
marks[1]=70
print(type(marks))
print(len(marks))
print(marks[8])

#slicing on list
print(marks[2:6])
print(marks[:])
print(marks[2:])
print(marks[:6])
print(marks[-1:-2])


# methods of list
#1] l.append( val)=> insert at the end 
#2] l.insert(index ,val)=> insert at index
#3] l.sort()=> arrange in ascending order
#4]l.reverse()=> reverse order
num=[1,2,3]
num.append(4)
print(num)
num.insert(2,10)
print(num)
num.sort(reverse=True)
print(num)
num.reverse()
print(num)


#looops in List
num=[1,2,3,4,10,5,6]
x=10
idx=0
for val in num:
    if(val==x):
        print(f"{x} found at index = {idx}")
        break
    idx+=1
        
```

```python
#dictionary => key value pair => keys are always unique
# mutable, unorderd
dict={"name":"Aarti",
      "cgpa":9.3,
     "subject":["math","science"],
      3.14:"PI"
    }
dict["cgpa"]=8.5
print(dict)
print(type(dict))

# we can access values through the keys in dictionary
print(dict["subject"])
print(dict["cgpa"])

#methods
#1] keys()=>retrurns all the keys of dict
#2] values()=>returns the values of dict
#3] items()=> returns all key value pair
#4] get()=> used to access value of the particular keys
#5] update()=> adds new item in dict

print(dict.keys())
print(dict.values())
print(dict.items())
print(dict.get("cgpa"))
dict.update({"city":"delhi"})
print(dict)
```

```python
#Q5
dict={
    "Aarti":70,
    "Ashwini":80,
    "sakshi":90,
    "kanchan":85
}
while True:
    print("\n-------MENU--------\n")
    print("A. Add Student")
    print("B. udate marks")
    print("C. search student")
    print("D. display all")
    print("E. Exit")

    ch= input("enter your choice").upper()

    if ch=='A':
        name=input("enter name to add")
        mark=int(input("enter marks"))
        dict[name]=mark
        print("student added")
    elif ch=='B':
        name=input("enter name update marks")
        mark=int(input("enter marks"))
        dict[name]=mark
        print("smarks udated")
    elif ch=='C':
        name=input("enter name to search")
        if name in dict:
            print(f"student found {dict[name]}")
    elif ch=='D':
        print(dict.items())
    elif ch=='E':
        print("exit")
        break
```
